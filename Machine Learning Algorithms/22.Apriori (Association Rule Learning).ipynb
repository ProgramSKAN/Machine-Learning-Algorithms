```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Data Preprocessing

```
#given the market dataset, recommand that people who bought productA may aslo like to buy productB. 
dataset = pd.read_csv('https://raw.githubusercontent.com/ProgramSKAN/Sample-Data-For-Machine-Learning/master/Market_Basket_Optimisation.csv', header = None)
transactions = []
for i in range(0, 7501):
  transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
```

## Training the Apriori model on the dataset

```
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
#ex:(3transaction/day *7)weeks/7501weeks  =min_support ie:products in the rules appear at least all point 3 percent of the time Than a minimum confidence 
#.Ex.get one product and get mostly likable product for free, so set min max length =2.Ex:with proct A on left we most likely to have product B on right atleat 20%>min_confidence
```

## Visualising the results

### Displaying the first results coming directly from the output of the apriori function

```
results = list(rules)
```

```
results
#[RelationRecord(items=frozenset({'chicken', 'light cream'}), support=0.004532728969470737, ordered_statistics=[OrderedStatistic(items_base=frozenset({'light cream'}),
#items_add=frozenset({'chicken'}), confidence=0.29059829059829057, lift=4.84395061728395)])

#ie:people who buy light cream they have high chances to buy chicken.that chance is measured bu confidence.
#if people buy light cream they have 29.059% chance to buy chicken.
#lift>3 : good
#the rule containing these 2 products appears in o.45% of the transactions.
```

### Putting the results well organised into a Pandas DataFrame

```
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```
