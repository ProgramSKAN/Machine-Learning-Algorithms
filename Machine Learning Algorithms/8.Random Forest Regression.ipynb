# **Random Forest Regression**

versions:Ensemble Learning,Gradient boosting
Random Forest is a version of Ensemble Learning.Ensemble learning is when you take multiple algorithms or the same algorithm multiple times and you put them together to make something much more powerful than the original.

step 1)pick at random K datapoints from training set
step 2)Build the Decision tree associated to these K datapoints
step 3)choose the number Ntree of trees you want to build and repeat step 1 & 2
step 4)for a new datapoint,make each one of your Ntree trees predict the value of Y to for the datapoint in the question and assign the new data point the average across all of the predicted values.

Ex:if we get 500 tree means we have 500 predictions for the value of y, then we take average of those.so increases accuracy.

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Import Dataset

```
dataset = pd.read_csv('https://raw.githubusercontent.com/ProgramSKAN/Sample-Data-For-Machine-Learning/master/Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

### Training the Random Forest Regression model on the whole dataset

```
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0) #n_estimators> number os trees.each tree is an estimator
regressor.fit(X, y)
```

### Predicting a new result

```
regressor.predict([[6.5]])
```

### Visualising the Random Forest Regression results (higher resolution)

```
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Previous Salary Truth or Bluff (Random Forest Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

here we see more steps than decision tree regression model because here we have more trees(10), so more splits.
