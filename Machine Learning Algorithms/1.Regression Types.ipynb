# **Simple Linear Regression**
In this we will predict a value of one variable Y (Dependent Variable) based on other variable X (Independent Variable).
Simple:Becasause it examins the relationship between two variables only.
Linear:Because, as the independent variable increases/decreases, the dependent variable increases/decreases in a linear fashion.

Ex: we want to predict the salary of an employee based on the number of years of experience.As number od years of experience goes up then the expectation is that the employee salary also goes up.

In this we use linear relationship with eqation
Y = m*X + B
Salary = m*(#years of experience) + B

Goal is to obtain the best value for coefficients m & B that best fit the data points.
m = slope of the line
B = Y-Intercept

Ex:m=4,B=1, so that HR decides on the salary of a new employee


# **Multiple Linear Regression**
It examins the relationship between more than two variables.Equation is

y = b0 + m1*x1 + m2*x2 +......+ mn*xn
x=independent variable
y=dependent variable

Ex: Assume in the above salary example, we need add one more parameter such as 'number of years of education' as a new independent variable.therefore

Salary = b0 + m1 *(#years of experience) + m2*(#years of education)

Goal is to find m1,m2,b0

# **Polynomial Regression**
It models the relationship as independent variable x and the dependent variable y as an nth degree polynomial in x.

y = b0 + b1*x + b2*x^2 + .... + bn*x^n

in simple model,in the scatter plot there will only a linear line which covers less data points (terrible fit).But by using polynomial regression the line plotted on scatter plot id curved and covering more data points(better fit).

As we increase the order of the polynomial, we increase the complexity of the model and capture much more complex relationships.Ofcourse number of parameters goes up as well.
