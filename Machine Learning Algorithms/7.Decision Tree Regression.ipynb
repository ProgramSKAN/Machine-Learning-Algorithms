# **Decision Tree Regression**
Decision trees are 2 types classification and regression.this is a Regressstion type Decision tree.

let, X1,X2 are independent variable and y is a dependent variable.we have a scatter plot between X1,X2.The algorithm do optimal splits of the datapoints into differnt sections, say 4 splits.The splits(leaf) is based on Infromation entropy.

After splitting, decision tres is drawn like X1<20? yes:no. yes X2<200 |no X2<170 ...soon.Now addding a new data point,this data point falls into any one split(terminal leaf).split is like an information.how does that information helps predicting value of y for new datapoint?

sol:take the average of each splited group, which is a predicted value of y.And splits/information increases then predictablility increases.the core of the algothithm is in bulding tree of conditions, so it is called decition tree.

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Importing the dataset

```
dataset = pd.read_csv('https://raw.githubusercontent.com/ProgramSKAN/Sample-Data-For-Machine-Learning/master/Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

### Training the Decision Tree Regression model on the whole dataset

```
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0) #fixing the seed using reandom_state parameter
regressor.fit(X, y)
```

### Predicting a new result

```
regressor.predict([[6.5]]) #not a good result.Descion tree is not good to use on single feature dataset.it is better for high dimensional dataset
```

### Visualising the Decision Tree Regression results (higher resolution)
lower resolution graph won't make sense because it just joins the points

```
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Previous Salary Truth or Bluff (Decision Tree Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

it is not well adapted to this 2D dataset.it is good for higher dimensional dataset.
